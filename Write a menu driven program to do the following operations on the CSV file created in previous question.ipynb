```python
import csv
con="y"
while con=="Y" or con=='y':
    print("Enter a number to execute corresponding operation: \n1. Insert record \n2. Display all records \n3. Search record by Roll No. \n4. Update record by Roll No. \n5. Delete record by Roll No.\n")
    n=int(input("Operation to be executed:"))
    print(" ")
    if n==1:
        with open("student3.csv",'a') as f:
            csvobj=csv.writer(f)
            Rno=int(input("Enter student roll number:"))
            Name=input("Enter student name:")
            Marks=eval(input("Enter student marks:"))
            l=[Rno,Name,Marks]
            csvobj.writerow(l)
            print("Record added!")
    if n==2:
        with open("student3.csv",'r') as g:
            a=csv.reader(g)
            for i in a:
                print(i)
    if n==3:
        with open("student3.csv",'r') as h:
            b=csv.reader(h)
            s=input("Enter roll number of record to be searched:")
            for i in b:
                if i[0]==s:
                    print(i)
    if n==4:
        with open("student3.csv",'r+') as p:
            c=list(csv.reader(p))
            r=input("Enter roll number of record to be modified:")
            for i in c:
                if i[0]==r:
                    i[1]=input("Enter new name:")
                    i[2]=eval(input("Enter new marks:"))
        with open("student3.csv",'w') as p:
            csvobj2=csv.writer(p)
            csvobj2.writerows(c)
            print("Record updated!")
    if n==5:
        with open("student3.csv",'r+') as t:
            d=csv.reader(t)
            u=input("Enter roll number of record to be deleted:")
            m=[]
            for i in d:
                if i[0]!=u:
                    m.append(i)
        with open("student3.csv",'w') as v:
            csvobj3=csv.writer(v)
            csvobj3.writerows(m)
        print("Record deleted!")
    con=input("Do you want to continue? Y / N: ")
print("Program ended!")
            

3
```
